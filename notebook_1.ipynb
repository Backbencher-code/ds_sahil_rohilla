# Handling missing market sentiment

After merging trader data with the Bitcoin market sentiment dataset, we observe that some trades
do not have an associated sentiment classification. This occurs because the Fear & Greed Index
is not available for every calendar date.

Since sentiment is a core explanatory variable in this analysis, trades without sentiment
information are excluded to ensure analytical integrity and avoid introducing noise.

```python
merged_df = merged_df.dropna(subset=['classification'])
```

```python
print(merged_df.shape)
merged_df['classification'].isna().sum()
```

```python
# A bitcoin only dataset
btc_df = merged_df[merged_df['Coin'] == 'BTC']
```

```python
closing_trades = merged_df[merged_df['Closed PnL'] != 0]
btc_closing_trades = btc_df[btc_df['Closed PnL'] != 0]
```

```python
merged_df['classification'].value_counts()
```

```python
# Profitability metrics : Profitability must be calculated only on closing trades.

closing_trades = merged_df[merged_df['Closed PnL'] !=0]
btc_closing_trades = closing_trades[closing_trades['Coin'] == 'BTC']
```

```python
# aVERAGE PnL by sentiment
avg_pnl_by_sentiment = (
    btc_closing_trades
    .groupby('classification')['Closed PnL']
    .mean()
    .sort_values(ascending=False)
)

avg_pnl_by_sentiment
```

```python
# Win rate by sentiment

btc_closing_trades['is_profitable'] = btc_closing_trades['Closed PnL'] > 0

win_rate_by_sentiment = (
    btc_closing_trades.groupby('classification')['is_profitable'].mean()
)

win_rate_by_sentiment
```

```python
# Loss magnitude

loss_trades = btc_closing_trades[btc_closing_trades['Closed PnL'] < 0]

avg_loss_by_sentiment = (
    loss_trades.groupby('classification')['Closed PnL'].mean()
)

avg_loss_by_sentiment
```

```python
# Tail risk
tail_risk_by_sentiment = (
    loss_trades.groupby('classification')['Closed PnL'].quantile(0.05)
)

tail_risk_by_sentiment
```

```python
# Trade volume by sentiment(BTC)
avg_trade_size_by_sentiment = (
    merged_df[merged_df['Coin'] == 'BTC'].groupby('classification')['Size USD'].mean()
)

avg_trade_size_by_sentiment
```

```python
# Trades per day
trades_per_day = (
merged_df[merged_df['Coin'] == 'BTC'].groupby(['trade_date', 'classification']).size()
.reset_index(name='trade_count')
)
```

```python
# average trades per day per sentiment
avg_trades_per_day = (
    trades_per_day.groupby('classification')['trade_count'].mean()
)

avg_trades_per_day
```

```python
# A summary table
summary_df = pd.DataFrame({
    'avg_pnl': avg_pnl_by_sentiment,
    'win_rate': win_rate_by_sentiment,
    'avg_loss': avg_loss_by_sentiment,
    'tail_risk_5pct': tail_risk_by_sentiment,
    'avg_trade_size_usd': avg_trade_size_by_sentiment,
    'avg_trades_per_day': avg_trades_per_day
})

summary_df
```

# Visualization and Pattern Validation

1. Average PnL by Market Sentiment

```python
plt.figure()
avg_pnl_by_sentiment.plot(kind='bar')
plt.title('Average BTC Trade PnL by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Closed PnL (USD)')
plt.tight_layout()
plt.savefig('outputs/avg_pnl_by_sentiment.png')
plt.show()
```

2. Win Rate by Market Sentiment

```python
plt.figure()
win_rate_by_sentiment.plot(kind='bar')
plt.title('Win Rate of BTC Trades by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Win Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.savefig('outputs/winRate_of_BTC_trades_by_market_sentiment.png')
plt.show()
```

3. Downside Risk (Average loss) by Sentiment

```python
plt.figure()
avg_loss_by_sentiment.plot(kind='bar')
plt.title('Average Loss per BTC Trade by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Loss (USD)')
plt.tight_layout()
plt.savefig('outputs/avg_loss_perBTC-Trade_by_sentiment.png')
plt.show()
```

4. Tail Risk

```python
plt.figure()
tail_risk_by_sentiment.plot(kind='bar')
plt.title('Tail Risk (5% Worst Losses) by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('5th Percentile Loss (USD)')
plt.tight_layout()
plt.savefig('outputs/tail_risk_by_sentiment.png')
plt.show()
```

5. Trading Activity

```python
plt.figure()
avg_trades_per_day.plot(kind='bar')
plt.title('Average Number of BTC Trades per Day by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Trades per Day')
plt.tight_layout()
plt.savefig('outputs/avgNo_BTC-trades_perDay_by_sentiment.png')
plt.show()
```

### Visual Pattern Validation

The visual analysis confirms that trader behavior and performance vary significantly across
market sentiment regimes. Fear-driven markets exhibit higher participation and stronger
risk-adjusted performance, while Greed-dominated periods show lower win rates and weaker returns,
suggesting overconfidence-driven trading.

Tail risk analysis reveals that extreme losses tend to cluster during Extreme Fear and Neutral
periods, highlighting the importance of strict risk management even when markets appear stable.
